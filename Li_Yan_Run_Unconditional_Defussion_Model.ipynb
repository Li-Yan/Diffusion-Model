```python
model_id = "li-yan/diffusion-aurora-256" #@param {type:"string"}
num_inference_steps = 100 #@param {type:"integer"}
```

```python
%pip install -qqU diffusers accelerate
```

```python
import torch

# Set device
device = (
    "mps"
    if torch.backends.mps.is_available()
    else "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
```

```python
from diffusers import DiffusionPipeline, DDIMScheduler

scheduler = DDIMScheduler.from_pretrained(model_id)
scheduler.set_timesteps(num_inference_steps=num_inference_steps)

pipeline = DiffusionPipeline.from_pretrained(model_id, scheduler=scheduler).to(device)
```

```python
images = pipeline(num_inference_steps=num_inference_steps).images
images[0]
```
